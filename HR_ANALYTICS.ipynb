```python
sns.countplot(x="Attrition",data=df)
plt.title("Distribution of Attrition")
plt.xlabel("Attrition")
plt.ylabel("Count")
plt.show()
```

```python
attrition_counts=df['Attrition'].value_counts()
total_employees=attrition_counts.sum()
attrition_percentages=attrition_counts/total_employees*100
print(attrition_percentages)
print(attrition_counts)
```

```python
attrition_counts=df['Attrition'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(attrition_counts,labels=attrition_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Distribution of Attrition')
plt.show()
```

```python
department_counts = df["Department"].value_counts()
total_employees = department_counts.sum()
department_percentages = (department_counts / total_employees) * 100
print("Percentage of employees in each department:")
print(department_percentages)
```

```python
department_counts = df["Department"].value_counts()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Employees by Department')
plt.show()
```

```python
jobrole_counts = df["JobRole"].value_counts()
total_employees = jobrole_counts.sum()
jobrole_percentages = (jobrole_counts / total_employees) * 100
```

```python
plt.figure(figsize=(8, 8))
plt.pie(jobrole_counts, labels=jobrole_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Employees by Job Role')
plt.show()
```

```python
attrition_by_department = df.groupby('Department')['Attrition'].value_counts().unstack()
attrition_by_department['AttritionRate'] = attrition_by_department['Yes'] / (attrition_by_department['Yes'] + attrition_by_department['No']) * 100
highest_attrition_department = attrition_by_department.sort_values(by='AttritionRate', ascending=False)
print(highest_attrition_department[['AttritionRate']])
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=highest_attrition_department.index, y=highest_attrition_department['AttritionRate'])
plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.show()
```
